# Imports

Now that the installation is done, Let me go forward to import the necessary modules. Followint keras and matplotlib for better visualization.

```python
import keras_ocr
import matplotlib.pyplot as plt
```

# Download Weights

Here, I had called the pipeline function from keras_ocr.pipeline class and stored it into the variable called as pipeline. Here, keras_ocr will automatically downlad the pretrained weights for the detector and recognizer and stored it into the varible called as pipeline.

```python
pipeline = keras_ocr.pipeline.Pipeline()
```

# Get Images

```python
images = [keras_ocr.tools.read(img) for img in ['/content/drive/MyDrive/Colab Notebooks/Image1.jpg', '/content/drive/MyDrive/Colab Notebooks/Image2.jpeg']
          ]
```

```python
len(images)
```

```python
plt.figure(figsize = (10, 20))
plt.imshow(images[0])
```

```python
plt.figure(figsize = (10, 20))
plt.imshow(images[1])
```

# Prediction

```python
prediction_groups = pipeline.recognize(images)
```

```python
fig, axs = plt.subplots(nrows=len(images), figsize=(10, 20))
for ax, image, predictions in zip(axs, images, prediction_groups):
  keras_ocr.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)
```
